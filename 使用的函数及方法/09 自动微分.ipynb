```python
'''
1.先正向传播，算预估值。
2.基于损失函数，和真实值，计算梯度。
3.结合权重更新公式，更新权重。
'''

import torch
#假设2行5列，全1矩阵。
x = torch.ones(2,5) 
print(f'x:{x}')

#假设2行3列，全0矩阵。
y = torch.zeros(2,3)
print(f'y:{y}')

#初始化权重和偏置
w = torch.randn(5,3,requires_grad=True)
print(f'w:{w}')

b = torch.randn(3,requires_grad=True)
print(f'b:{b}')

#前向传播，算预测值
z = torch.matmul(x,w) + b
print(f'z:{z}')

#定义损失函数
criterion = torch.nn.MSELoss()#nn:neural netwoek:神经网络
loss = criterion(z,y) 

#自动微分，结合反向传播，更新权重
loss.sum().backward()

print(f'w梯度:{w.grad} b的梯度:{b.grad}')

```
