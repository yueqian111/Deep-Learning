```python
"""
=========================================
附录：常见损失函数 (Loss Function) 总结备忘录
=========================================

【常见损失函数选择】
1. 分类问题：
   - 多分类交叉熵损失：CrossEntropyLoss
   - 二分类交叉熵损失：BCELoss

2. 回归问题：
   - MAE: Mean Absolute Error, 平均绝对误差。
   - MSE: Mean Squared Error, 均方误差。
   - Smooth L1: 结合了 MAE 和 MSE 两个的特点做的升级、优化。

-----------------------------------------
【深入理解：多分类交叉熵损失 (CrossEntropyLoss)】

1. 设计思路 (数学公式)：
   Loss = - Σ y * log(S(f(x)))

2. 变量简单记忆：
   * x:       样本
   * f(x):    加权求和（网络的原始输出）
   * S(f(x)): 处理后的概率（经过 Softmax 映射后的概率分布）
   * y:       样本 x 属于某一个类别的“真实概率”（通常是 One-Hot 编码）

3. 大白话解释：
   损失函数的目标 = 让“正确类别”对应概率的对数最大化（加上负号就是最小化损失）。简而言之，就是拼命拉高正确答案的预测概率。

4. ⚠️ 核心细节 (新手极易踩坑点)：
   CrossEntropyLoss 底层源码实际上等于 = Softmax() + 负对数似然损失计算。
   结论：如果训练时使用了 CrossEntropyLoss 这个损失函数，那么【神经网络的输出层就绝对不要额外调用 softmax() 激活函数了】，否则相当于做了两次概率映射，会导致模型难以收敛！
"""
import torch
import torch.nn as nn

#演示多分类交叉熵损失
def dm01():
    #y_true = torch.tensor(data=[0,1,0],[1,0,0]],dtype=torch.float)
    #手动创建真实值
    y_true = torch.tensor([1,2])
    #手动创建预测值
    y_pred = torch.tensor([[0.1,0.8,0.1],[0.7,0.2,0.1]],requires_grad=True,dtype=torch.float)
    #创建多分类交叉损失函数
    criterion = nn.CrossEntropyLoss()
    loss = criterion(y_pred,y_true)
    print(f'损失值：{loss}')
dm01()
```
